```python
# Import Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Data Collection
dataset=pd.read_csv("Insurance-Data.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
# Input and Output Split
independent=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent=sc.fit_transform(independent)
```

```python
# In Regression GridSearchCv no need of split train and test set
# Using GridSearchCV and Creating Model
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'C': [ 10, 100, 1000, 2000, 3000],
			  'gamma': ['auto','scale'], 
			  'kernel': ['rbf','sigmoid','linear']} 
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3, n_jobs = -1) 
grid.fit(independent,dependent) 
```

```python
y_pred=grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(y_pred,dependent)
```

```python
# Print best parameter after tuning
re=grid.cv_results_
print("R_score value for best parameter{}:".format(grid.best_params_), r_score)
```

```python
table=pd.DataFrame.from_dict(re)
```
